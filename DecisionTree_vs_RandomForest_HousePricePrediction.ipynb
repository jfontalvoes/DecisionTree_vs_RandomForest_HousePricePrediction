### MAE (Mean Absolute Error)

Se eval煤a el desempe帽o del modelo utilizando el Mean Absolute Error (MAE), una m茅trica que mide el promedio de las diferencias absolutas entre los valores reales y las predicciones.
Un MAE bajo indica que, en promedio, las predicciones est谩n cerca de los valores reales.

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)

valorMAE = mean_absolute_error(y, predicted_home_prices)
print('El MAE es de: ', valorMAE)
```

En este caso, el valor de MAE obtenido es de aproximadamente 1115.75, lo cual representa un error extremadamente bajo considerando las escalas de precio manejadas en el dataset.
Este resultado sugiere la posibilidad de overfitting, ya que es muy poco com煤n obtener un error tan peque帽o en un problema de regresi贸n con datos reales.
Cuando se presenta overfitting, el modelo aprende demasiado bien los datos de entrenamiento, pero pierde capacidad de generalizaci贸n sobre nuevos datos.

##  Divisi贸n de Datos y Validaci贸n del Modelo
Para evaluar correctamente el desempe帽o del modelo y evitar sobreajuste, se divide el dataset en dos subconjuntos:

* Conjunto de entrenamiento: utilizado para ajustar los par谩metros del modelo.

* Conjunto de validaci贸n: empleado para medir la capacidad predictiva del modelo sobre datos no vistos.

Posteriormente, se entrena un Decision Tree Regressor sobre el conjunto de entrenamiento y se eval煤a su desempe帽o calculando el Mean Absolute Error (MAE) sobre el conjunto de validaci贸n.

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_y)

val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

En este caso, se obtuvo un MAE de aproximadamente 269,609.90, considerablemente m谩s alto que el error sobre los datos completos, lo cual confirma que el modelo original presentaba overfitting.
Este resultado valida la importancia de separar los datos y realizar una prueba honesta de rendimiento.

## Comparaci贸n de Tama帽os de rbol
Se prueba c贸mo cambia el error al ajustar el n煤mero m谩ximo de nodos hoja (max_leaf_nodes) en el 谩rbol.

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

A medida que se permite que el 谩rbol tenga m谩s nodos, el MAE baja bastante, pasando de m谩s de 385K a unos 260K.

Esto muestra que un 谩rbol muy peque帽o se queda corto, y uno m谩s grande mejora la precisi贸n, aunque hay que cuidar no pasarse para evitar overfitting.

#  Entrenamiento ML de Random Forest
Despu茅s de probar con el 谩rbol de decisi贸n, se busca mejorar los resultados usando Random Forest, que en vez de hacer un solo 谩rbol, construye varios y promedia sus predicciones.
Esto suele reducir el error porque disminuye la varianza del modelo y evita que se sobreentrene tan f谩cil como un solo 谩rbol.

La idea es comparar ambos modelos con el MAE y ver si realmente se gana precisi贸n.

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print('El MAE es:',mean_absolute_error(val_y, melb_preds))
```

### B煤squeda de Hiperpar谩metros con GridSearchCV
Para sacarle mejor provecho al Random Forest, se ajustan varios hiperpar谩metros como la cantidad de 谩rboles, profundidad m谩xima y tama帽o m铆nimo de muestras por hoja.

Se usa GridSearchCV, que prueba todas las combinaciones posibles de los valores definidos en param_grid, usando validaci贸n cruzada, y se queda con la que d茅 el MAE m谩s bajo. As铆 se optimiza el modelo sin tener que ir tanteando valores manualmente.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=1)


param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Crear modelo base
rf = RandomForestRegressor(random_state=1)

# Configurar GridSearch
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_mean_absolute_error')


grid_search.fit(X_train, y_train)

print("Mejores par谩metros:")
print(grid_search.best_params_)
print("Mejor MAE:", -grid_search.best_score_)
```

###  Entrenar el Modelo RF Final con los Mejores Par谩metros
Despu茅s de encontrar la mejor combinaci贸n con GridSearchCV, se entrena un nuevo Random Forest usando esos hiperpar谩metros optimizados.

Se ajusta el modelo sobre el conjunto de entrenamiento y se calcula el MAE final sobre el conjunto de validaci贸n para ver cu谩nto mejor贸 respecto a las pruebas anteriores.

```python
# Sacar los mejores par谩metros
best_params = grid_search.best_params_

final_model = RandomForestRegressor(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=1
)

final_model.fit(X_train, y_train)
final_preds = final_model.predict(X_valid)
mae_final = mean_absolute_error(y_valid, final_preds)
print("MAE final es:", mae_final)
```
